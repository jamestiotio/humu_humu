```python
import numpy as np
#top_five_patterns is a top 5 list of the most probable order of tags

#placeholder
tags=np.array([0,1,2,3,4,5,6])
pi=np.array([[]])
a=np.array([[]])

#following HMM notes 1 and assuming tags list already contains START and STOP
tag_stop=len(tags)-1
tag_start=0

#this stores the top five order of tags 
top_five_patterns=np.array([
    [tag_stop],
    [tag_stop],
    [tag_stop],
    [tag_stop],
    [tag_stop]
])
def viterbi_backward_top(n,top_five_patterns):
    total_scores=[]
    total_tag_patterns=[]
    #we don't know where the top 5 patterns will come from so calculate from the previous top 5 and sort for the score
    for candidate in top_five_patterns:
        yn=candidate[0]
        scores=[]
        
        #calculate score for every tag
        for tag_index in range(len(tags)):
            score=pi[n,tag_index]*a[tag_index,yn]
            scores.append(score)
        #sort by the score
        tag_sorted_indices = np.array(scores).argsort()

        tags_sorted = tags[tag_sorted_indices[::-1]]
        # tags_sorted = [x for _,x in sorted(zip(scores,tags))]
        scores.sort(reverse=True)
    
        #take the top 5 tags and scores
        tags_sorted =tags_sorted[:5]
        scores =scores[:5] 
        total_scores.extend(scores)
        total_tag_patterns.extend(tags_sorted)
    
    #id list to be sorted by score so we can keep track of the candidate used for the tag
    candidate_ids=np.concatenate((np.zeros(5),np.ones(5),np.ones(5)*2,np.ones(5)*3,np.ones(5)*4))
    #sort by the score again  and take top 5
    total_tag_patterns = [x for _,x in sorted(zip(total_scores,total_tag_patterns))][-5:]
    candidate_ids = [x for _,x in sorted(zip(total_scores,candidate_ids))][-5:]

    # form the new top 5 patterns
    for i in range(5):
        prev_pattern=top_five_patterns[candidate_ids[i]]
        new_token=tags[total_tag_patterns[i]]
        top_five_patterns[i]=[new_token]+prev_pattern

            
    if n>0:
        # continue going backwards
        return viterbi_backward_top(n-1,top_five_patterns)
    else:
        # done
        return top_five_patterns
```
